```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Waste Management': [85, 70, 95, 60, 75, 80, 55, 90, 65, 50],
    'Conservation Efforts': [90, 60, 85, 70, 75, 90, 50, 95, 65, 40],
    'Local Biodiversity': [80, 55, 75, 65, 70, 85, 60, 95, 60, 45],
    'Water Quality': [90, 75, 80, 70, 85, 88, 60, 95, 80, 65],
    'Sustainability Rating': [85, 65, 82, 66, 73, 83, 53, 96, 62, 42]
}

df = pd.DataFrame(data)

X = df.drop(columns=['Sustainability Rating'])
y = df['Sustainability Rating']

def train_and_evaluate(split_ratio):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-split_ratio, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return model, r2, y_pred, X_test, y_test

model_80_20, r2_80_20, predictions_80_20, X_test_80, y_test_80 = train_and_evaluate(0.8)
model_70_30, r2_70_30, predictions_70_30, X_test_70, y_test_70 = train_and_evaluate(0.7)

print("R² Score (80:20 split):", r2_80_20)
print("R² Score (70:30 split):", r2_70_30)

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

features = correlation_matrix['Sustainability Rating'].abs().sort_values(ascending=False).index[1:]
print("Best subset of features based on correlation:\n", features)

X_best = X[features]
model_best = LinearRegression()
model_best.fit(X_best, y)

print("Best Model Parameters:", model_best.coef_)

random_indices = np.random.choice(X_best.index, 5, replace=False)
print("Predictions for random test examples:")
for idx in random_indices:
    sample = X_best.loc[[idx]]  # Ensure it is a DataFrame
    print(f"Actual: {y[idx]}, Predicted: {model_best.predict(sample)[0]}")

train_mse = mean_squared_error(y, model_best.predict(X_best))
print("Training MSE:", train_mse)

test_mse = mean_squared_error(y_test_70, model_best.predict(X_test_70[features]))
if train_mse > test_mse:
    print("The model is underfitting.")
elif train_mse < test_mse:
    print("The model is overfitting.")
else:
    print("The model is well-fitted.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Beach Name': ['Beach A', 'Beach B', 'Beach C', 'Beach D', 'Beach E',
                   'Beach F', 'Beach G', 'Beach H', 'Beach I', 'Beach J'],
    'Waste Management': [85, 70, 95, 60, 75, 80, 55, 90, 65, 50],
    'Conservation Efforts': [90, 60, 85, 70, 75, 90, 50, 95, 65, 40],
    'Local Biodiversity': [80, 55, 75, 65, 70, 85, 50, 95, 60, 45],
    'Water Quality': [90, 75, 80, 70, 85, 88, 60, 95, 80, 65],
    'Sustainability Rating': [85, 65, 82, 66, 73, 83, 53, 96, 62, 42]
}

df = pd.DataFrame(data)
df.drop(columns=['Beach Name'], inplace=True)
threshold = 70
df['Sustainable'] = np.where(df['Sustainability Rating'] >= threshold, 1, 0)

X = df.drop(columns=['Sustainability Rating', 'Sustainable'])
y = df['Sustainable']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Not Sustainable', 'Sustainable'], 
            yticklabels=['Not Sustainable', 'Sustainable'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


data = {
    'Beach Name': ['Beach A', 'Beach B', 'Beach C', 'Beach D', 'Beach E',
                   'Beach F', 'Beach G', 'Beach H', 'Beach I', 'Beach J'],
    'Waste Management': [85, 70, 95, 60, 75, 80, 55, 90, 65, 50],
    'Conservation Efforts': [90, 60, 85, 70, 75, 90, 50, 95, 65, 40],
    'Local Biodiversity': [80, 55, 75, 65, 70, 85, 60, 95, 60, 45],
    'Water Quality': [90, 75, 80, 70, 85, 88, 60, 95, 80, 65],
    'Sustainability Rating': [1, 0, 1, 0, 0, 1, 0, 1, 0, 0]  # Binary target
}

df = pd.DataFrame(data)


X = df.drop(columns=['Beach Name', 'Sustainability Rating'])
y = df['Sustainability Rating']

def evaluate_model(split_ratio):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-split_ratio, random_state=42, stratify=y)
    
    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy


accuracy_80_20 = evaluate_model(0.2)
accuracy_70_30 = evaluate_model(0.3)

print("Accuracy (80:20 split):", accuracy_80_20)
print("Accuracy (70:30 split):", accuracy_70_30)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

def evaluate_with_penalty(penalty, l1_ratio=None):
    if penalty in ['l1', 'elasticnet']:
        solver = 'saga'
    else:
        solver = 'lbfgs'

    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    model = LogisticRegression(max_iter=5000, penalty=penalty, solver=solver, l1_ratio=l1_ratio)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy


accuracies = {}
accuracies['l1'] = evaluate_with_penalty('l1')
print(f"Accuracy with penalty 'l1': {accuracies['l1']}")

accuracies['l2'] = evaluate_with_penalty('l2')
print(f"Accuracy with penalty 'l2': {accuracies['l2']}")

accuracies['none'] = evaluate_with_penalty(None)
print(f"Accuracy with penalty 'none': {accuracies['none']}")

accuracies['elasticnet'] = evaluate_with_penalty('elasticnet', l1_ratio=0.5)
print(f"Accuracy with penalty 'elasticnet': {accuracies['elasticnet']}")
```
